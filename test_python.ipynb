## Exercice 3 : tableau de bord

Pour avoir une vision synthétique des informations, une nouvelle version du programme est mise en oeuvre. Le responsable d'exploitation souhaite disposer d'un tableau de bord affichant la température moyenne mesurée dans la journée, la température maxi et la température mini.

Vous devrez obtenir l'affichage ci-dessous:
```
Tableau de bord
---------------------------------------------------------
                moy min max
chambrefroide01 3.4 2.5 6.4
chambrefroide02 2.8 2.5 4.1
chambrefroide03 2.7 2.5 3.6
```

````python
# à retravailler un peu plus d'un point de vue sur l'indentation et sur les cas où les données sont vides
import csv
from collections import defaultdict

# Affichage de l'en-tête du programme
print("Tableau de bord")
print('-' * 57)
print('                moy min max')

# Utilisation de defaultdict stocker les températures par chambre froide
chambres_froides = defaultdict(lambda: {'sum': 0, 'count': 0, 'min': float('inf'), 'max': float('-inf')})

# Ouverture du fichier de données
with open('traiteur.csv') as csv_file:
    contenu_csv = csv.reader(csv_file, delimiter=';')
    
    # Parcours des données du fichier
    for champ in contenu_csv:
        if len(champ) < 4:
            # Si la ligne n'a pas assez de colonnes, on l'ignore
            continue

        chambre = champ[2]
        #l'écrire explicitement + Faire la vérification des données (si la variable est un numérique, alors on fait le traitement.)
        try:
            temperature = float(champ[3].replace(',', '.'))
            chambres_froides[chambre]['sum'] += temperature
            chambres_froides[chambre]['count'] += 1
            chambres_froides[chambre]['min'] = min(chambres_froides[chambre]['min'], temperature)
            chambres_froides[chambre]['max'] = max(chambres_froides[chambre]['max'], temperature)
        except ValueError:
            continue

# Calcul et affichage des résultats pour chaque chambre froide
for chambre, data in chambres_froides.items():
    if data['count'] > 0:
        moyenne = data['sum'] / data['count']
        print(f'{chambre} {moyenne:.1f} {data["min"]:.1f} {data["max"]:.1f}')
````

````python
#Nouvelle version script Exo3
import csv
from collections import defaultdict


def afficher_entete():
    print("Tableau de bord")
    print('-' * 57)
    print(f'{"Chambre":<13} {"Moy":>7} {"Min":>5} {"Max":>5}')


def traiter_ligne(champ, chambres_froides):
    if len(champ) < 4:
        return  # Ligne incomplète, on ignore

    chambre = champ[2].strip()
    temperature_str = champ[3].strip()

    if not chambre or not temperature_str:
        return  # Si une donnée clé est vide, on ignore

    try:
        temperature = float(temperature_str.replace(',', '.'))
        data = chambres_froides[chambre]
        data['sum'] += temperature
        data['count'] += 1
        data['min'] = min(data['min'], temperature)
        data['max'] = max(data['max'], temperature)
    except ValueError:
        pass  # Ignorer les températures non valides


def calcul_et_affichage(chambres_froides):
    for chambre, data in chambres_froides.items():
        if data['count'] > 0:
            moyenne = data['sum'] / data['count']
            print(f'{chambre:<15} {moyenne:>5.1f} {data["min"]:>5.1f} {data["max"]:>5.1f}')
        else:
            print(f'{chambre:<15} Aucune donnée disponible')

def main():
    """Point d'entrée principal du programme."""
    afficher_entete()

    chambres_froides = defaultdict(lambda: {'sum': 0, 'count': 0, 'min': float('inf'), 'max': float('-inf')})
    
    try:
        with open('traiteur.csv') as csv_file:
            for champ in csv.reader(csv_file, delimiter=';'):
                traiter_ligne(champ, chambres_froides)
    except FileNotFoundError:
        print("Erreur : Le fichier 'traiteur.csv' est introuvable.")
        return

    calcul_et_affichage(chambres_froides)

if __name__ == "__main__":
    main()
````
